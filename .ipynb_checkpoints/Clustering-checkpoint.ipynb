```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import pickle
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import RobustScaler
from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.cluster import DBSCAN 
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import linkage,dendrogram
```

```python
data = pd.read_csv(r'../data/CURATED/df_bien.csv', encoding="UTF8", sep=',')
df_bien = data.copy()
```

<div style="background-color: RGB(51,165,182);" >
<h2 style="margin: auto; padding: 20px; color:#fff; ">Clustering Analyzis</h2>
</div>

```python
df = data.copy()
df['Surface reelle bati'].sort_values()
```

```python
for col in ['Valeur fonciere','Surface reelle bati', 'Prix m2']:
    df[col]=np.log10(df[col])
scaler = RobustScaler()
XX=scaler.fit_transform(df[['Valeur fonciere','Surface reelle bati','Nombre pieces principales','Prix m2']])
```

```python
kmean = KMeans(n_clusters=7, n_init=10, init='k-means++')
yy_kmeans= kmean.fit_predict(XX)

for i in range (1,4):
    fig = plt.figure(figsize=(14, 6))
    plt.scatter(XX[yy_kmeans == 0, 0], XX[yy_kmeans == 0, i], s = 4, c = 'red', label = 'Cluster 1')
    plt.scatter(XX[yy_kmeans == 1, 0], XX[yy_kmeans == 1, i], s = 4, c = 'blue', label = 'Cluster 2')
    plt.scatter(XX[yy_kmeans == 2, 0], XX[yy_kmeans == 2, i], s = 4, c = 'green', label = 'Cluster 3')
    plt.scatter(XX[yy_kmeans == 3, 0], XX[yy_kmeans == 3, i], s = 4, c = 'cyan', label = 'Cluster 4')
    plt.scatter(XX[yy_kmeans == 4, 0], XX[yy_kmeans == 4, i], s = 4, c = 'magenta', label = 'Cluster 5')
    plt.scatter(XX[yy_kmeans == 5, 0], XX[yy_kmeans == 5, i], s = 4, c = 'black', label = 'Cluster 6')
    plt.scatter(XX[yy_kmeans == 6, 0], XX[yy_kmeans == 6, i], s = 4, c = 'orange', label = 'Cluster 7')
    plt.scatter(kmean.cluster_centers_[:, 0], kmean.cluster_centers_[:, i], s = 50, c = 'yellow', label = 'Centroids')
    plt.title('Clusters of customers')
    plt.legend()
    plt.show()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">1 - Kmeans</h3>
</div>

```python
scaler = RobustScaler()
X=scaler.fit_transform(df_bien[['Valeur fonciere','Surface reelle bati','Nombre pieces principales','Prix m2']])
Y=pd.DataFrame(X, columns=['Valeur fonciere','Surface reelle bati','Nombre pieces principales','Prix m2'])
```

```python
kmeans = KMeans(n_clusters=7, n_init=10, init='k-means++')
y_kmeans= kmeans.fit_predict(X)
```

```python
kmeans.labels_
```

```python
for i in range (1,4):
    fig = plt.figure(figsize=(14, 6))
    plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, i], s = 10, c = 'red', label = 'Cluster 1')
    plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, i], s = 10, c = 'blue', label = 'Cluster 2')
    plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, i], s = 10, c = 'green', label = 'Cluster 3')
    plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, i], s = 10, c = 'cyan', label = 'Cluster 4')
    plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, i], s = 10, c = 'magenta', label = 'Cluster 5')
    plt.scatter(X[y_kmeans == 5, 0], X[y_kmeans == 5, i], s = 10, c = 'black', label = 'Cluster 6')
    plt.scatter(X[y_kmeans == 6, 0], X[y_kmeans == 6, i], s = 10, c = 'orange', label = 'Cluster 7')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, i], s = 50, c = 'yellow', label = 'Centroids')
    plt.title('Clusters of customers')
    plt.legend()
    plt.show()
```

```python
for i in range (0,3):
    fig = plt.figure(figsize=(14, 6))
    plt.scatter(X[y_kmeans == 0, 3], X[y_kmeans == 0, i], s = 10, c = 'red', label = 'Cluster 1')
    plt.scatter(X[y_kmeans == 1, 3], X[y_kmeans == 1, i], s = 10, c = 'blue', label = 'Cluster 2')
    plt.scatter(X[y_kmeans == 2, 3], X[y_kmeans == 2, i], s = 10, c = 'green', label = 'Cluster 3')
    plt.scatter(X[y_kmeans == 3, 3], X[y_kmeans == 3, i], s = 10, c = 'cyan', label = 'Cluster 4')
    plt.scatter(X[y_kmeans == 4, 3], X[y_kmeans == 4, i], s = 10, c = 'magenta', label = 'Cluster 5')
    plt.scatter(X[y_kmeans == 5, 3], X[y_kmeans == 5, i], s = 10, c = 'black', label = 'Cluster 6')
    plt.scatter(X[y_kmeans == 6, 3], X[y_kmeans == 6, i], s = 10, c = 'orange', label = 'Cluster 7')
    plt.scatter(kmeans.cluster_centers_[:, 3], kmeans.cluster_centers_[:, i], s = 50, c = 'yellow', label = 'Centroids')
    plt.title('Clusters of customers')
    plt.legend()
    plt.show()
```

```python
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
clusters = [4,5,6,7]
cluster_centers = {}

for c in clusters:
    kms = KMeans(n_clusters=c, init='k-means++').fit(Y)
    Y['cluster' + str(c)] = kms.labels_
    cluster_centers[str(c)] = kms.cluster_centers_
    
```

```python
plt.figure(figsize=(20,15))
for i, c in enumerate(clusters):
    plt.subplot(2,2,i+1)
    sns.scatterplot(Y['Valeur fonciere'],Y['Surface reelle bati'], Y['cluster' + str(c)], s=50, palette=sns.color_palette("hls", c))
    sns.scatterplot(cluster_centers[str(c)][:,0], cluster_centers[str(c)][:,1], color='black', s=100)
    plt.title('Number of clusters: ' + str(c))
    
plt.show()
```

```python
plt.figure(figsize=(20,15))
for i, c in enumerate(clusters):
    plt.subplot(2,2,i+1)
    sns.scatterplot(Y['Valeur fonciere'],Y['Nombre pieces principales'], Y['cluster' + str(c)], s=50, palette=sns.color_palette("hls", c))
    sns.scatterplot(cluster_centers[str(c)][:,0], cluster_centers[str(c)][:,2], color='black', s=100)
    plt.title('Number of clusters: ' + str(c))
    
plt.show()
```

```python
plt.figure(figsize=(20,15))
for i, c in enumerate(clusters):
    plt.subplot(2,2,i+1)
    sns.scatterplot(Y['Valeur fonciere'],Y['Prix m2'], Y['cluster' + str(c)], s=50, palette=sns.color_palette("hls", c))
    sns.scatterplot(cluster_centers[str(c)][:,0], cluster_centers[str(c)][:,3], color='black', s=100)
    plt.title('Number of clusters: ' + str(c))
    
plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))
ax = Axes3D(fig, rect=[0, 0, .99, 1], elev=20, azim=210)
ax.scatter(Y['Valeur fonciere'],
           Y['Surface reelle bati'],
           Y['Nombre pieces principales'],
           c=Y['cluster5'],
           s=15,  cmap=plt.cm.Set1)

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Valeur fonciere')
ax.set_ylabel('Surface reelle bati')
ax.set_zlabel('nb de piece')
ax.set_title('3D view of K-Means 5 clusters')
ax.dist = 12

plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))
ax = Axes3D(fig, rect=[0, 0, .99, 1], elev=20, azim=210)
ax.scatter(Y['Prix m2'],
           Y['Surface reelle bati'],
           Y['Nombre pieces principales'],
           c=Y['cluster5'],
           s=15, cmap=plt.cm.Set1)

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Prix m²')
ax.set_ylabel('Surface reelle bati')
ax.set_zlabel('nb de piece')
ax.set_title('3D view of K-Means 5 clusters')
ax.dist = 12

plt.show()
```

<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">2 - DBSCAN </h3>
</div>

```python
db = DBSCAN(eps=0.5, min_samples=10).fit(Y)
```

```python
Y['Labels_db'] = db.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(Y['Valeur fonciere'],Y['Nombre pieces principales'], c=Y['Labels_db'], 
                cmap=plt.cm.Set1)
plt.title('DBSCAN with epsilon 11, min samples 6')
plt.show()
```

```python
dbs = DBSCAN(eps=2.5, min_samples=16).fit(Y)
```

```python
Y['Labels_db'] = dbs.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(Y['Valeur fonciere'],Y['Surface reelle bati'], c=Y['Labels_db'], cmap=plt.cm.Set1)
plt.title('DBSCAN with epsilon 5, min samples 6')
plt.show()
```

```python
YY=pd.DataFrame(XX, columns=['Valeur fonciere','Surface reelle bati','Nombre pieces principales','Prix m2'])
```

```python
dbss = DBSCAN(eps=2.5, min_samples=16).fit(YY)
```

```python
YY['Labels_db'] = dbss.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(YY['Valeur fonciere'], YY['Surface reelle bati'], c=YY['Labels_db'], cmap=plt.cm.Set1, s=5)
plt.title('DBSCAN with epsilon 5, min samples 6')
plt.show()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">3 - MeanShift </h3>
</div>

```python
bandwidth = estimate_bandwidth(Y, quantile=0.1)
ms = MeanShift(bandwidth).fit(Y)

Y['Labels_M'] = ms.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(Y['Valeur fonciere'],Y['Surface reelle bati'], hue=Y['Labels_M'], 
                palette=sns.color_palette('hls', np.unique(ms.labels_).shape[0]))
plt.plot()
plt.title('MeanShift')
plt.show()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">4 - Hierarchical & AgglomerativeClustering </h3>
</div>

```python
agglom = AgglomerativeClustering(n_clusters=5, linkage='average').fit(Y)

Y['Labels_Agg'] = agglom.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(Y[0], Y[1], hue=Y['Labels_Agg'], 
                palette=sns.color_palette('hls', 5))
plt.title('Agglomerative with 5 Clusters')
plt.show()
```
